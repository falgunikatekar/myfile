```python
import cv2
import matplotlib.pyplot as plt

# Load the pre-trained Haar Cascade classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Read an image
image = cv2.imread('C:/Users/ASUS/OneDrive/Desktop/falguni.jpg')  # Replace 'face.jpg' with the path to your image

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 3)

# Display the output image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt
# Load image
image_path = "C:/Users/ASUS/OneDrive/Desktop/falguni.jpg"  # Replace with your image path
image = cv2.imread(image_path)

# Analyze emotion
result = DeepFace.analyze(image, actions=['emotion'])

# Display results
print("Detected Emotion:", result[0]['dominant_emotion'])
if result[0]['dominant_emotion'] == "neutral":
    print("\"Happiness is not a destination, it's a way of life.\" Take a deep breath")
elif result[0]['dominant_emotion'] == "happy":
    print("\"Happiness is when what you think, what you say, and what you do are in harmony.\"")
elif result[0]['dominant_emotion'] == "fear":
    print("Happiness is not the absence of fear, but the ability to face it with hope.")
elif result[0]['dominant_emotion'] == "sad":
    print("Sadness gives depth. Happiness gives height. Sadness gives roots. Happiness gives branches. ")

# Show image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
